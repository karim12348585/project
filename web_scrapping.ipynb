```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sqlalchemy import create_engine
import requests
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import logging

# Configuration
CONFIG = {
    'data_sources': [
        {
            'type': 'sql',
            'connection_string': 'Driver={SQL Server};Server=your_server;Database=your_database;UID=your_user;PWD=your_password;',
            'query': 'SELECT * FROM your_table'
        },
        {
            'type': 'api',
            'url': 'https://api.example.com/data',
            'params': {'key': 'value'}
        },
        {
            'type': 'csv',
            'filepath': '/path/to/your/file.csv'
        }
    ],
    'anomaly_detection': {
        'method': 'isolation_forest',
        'parameters': {'n_estimators': 100, 'contamination': 0.01}
    },
    'alerting': {
        'email': {
            'enabled': True,
            'smtp_server': 'smtp.example.com',
            'port': 587,
            'login': 'your_email@example.com',
            'password': 'your_password',
            'from_email': 'your_email@example.com',
            'to_email': 'recipient@example.com'
        }
    },
    'logging': {
        'level': logging.INFO,
        'filename': 'anomaly_detection.log'
    }
}

# Setup logging
logging.basicConfig(level=CONFIG['logging']['level'], filename=CONFIG['logging']['filename'],
                    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

# Data Extraction
def extract_data(source):
    if source['type'] == 'sql':
        engine = create_engine(source['connection_string'])
        return pd.read_sql(source['query'], engine)
    elif source['type'] == 'api':
        response = requests.get(source['url'], params=source['params'])
        response.raise_for_status()
        return pd.DataFrame(response.json())
    elif source['type'] == 'csv':
        return pd.read_csv(source['filepath'])
    else:
        raise ValueError("Unsupported data source type")

# Data Cleaning and Preparation
def clean_data(df):
    # Add data cleaning steps as necessary
    df = df.dropna()
    return df

# Anomaly Detection
def detect_anomalies(df, method, parameters):
    if method == 'isolation_forest':
        model = IsolationForest(**parameters)
        df['anomaly'] = model.fit_predict(df)
        anomalies = df[df['anomaly'] == -1]
        return anomalies
    else:
        raise ValueError("Unsupported anomaly detection method")

# Alerting
def send_email_alert(subject, body, config):
    if not config['enabled']:
        return
    msg = MIMEMultipart()
    msg['From'] = config['from_email']
    msg['To'] = config['to_email']
    msg['Subject'] = subject

    msg.attach(MIMEText(body, 'plain'))

    server = smtplib.SMTP(config['smtp_server'], config['port'])
    server.starttls()
    server.login(config['login'], config['password'])
    server.sendmail(config['from_email'], config['to_email'], msg.as_string())
    server.quit()

def main():
    logger.info("Starting anomaly detection script")
    for source in CONFIG['data_sources']:
        try:
            logger.info(f"Extracting data from {source['type']} source")
            df = extract_data(source)
            df = clean_data(df)
            anomalies = detect_anomalies(df, CONFIG['anomaly_detection']['method'], CONFIG['anomaly_detection']['parameters'])
            if not anomalies.empty:
                logger.warning(f"Anomalies detected: {anomalies}")
                send_email_alert("Data Anomaly Detected", anomalies.to_string(), CONFIG['alerting']['email'])
        except Exception as e:
            logger.error(f"Error processing {source['type']} source: {e}")

if __name__ == "__main__":
    main()
```
